```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
algos_labels = ['CBS', 'CBS+PC', 'CBS+DS', 'CBS+H']
```

```python
def represent(min_agents, max_agents, results, ylabel, title, ax):
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    agents = range(min_agents, max_agents + 1)
    for values, label in zip(results, algos_labels):
        ax.scatter(agents, values, label=label)
        ax.plot(agents, values)
    ax.set_xlabel('Number of agents')
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_size_inches(18.5, 10.5)
represent(1, 3, [[1, 2, 3], [2, 2, 2], [3, 2, 1], [5, 4, 5]], 'Success rate', 'den520 map', ax1)
plt.show()
```
